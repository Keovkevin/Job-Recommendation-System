```python
df2=pd.read_csv("dice_com-job_us_sample.csv")
print(df2.head())
jobs=[]
for job_title  in df2.jobtitle:
    if(job_title.lower() not in jobs):
        jobs.append(job_title)
#print(jobs)
job_description=np.asarray(df2.loc[:,"jobdescription"])
print(len(job_description[0:5]))
```

```python
#Analysing the text corpus
#Given a text corpus, performs NER
def NER(textcorpus):
    tokenized=nltk.word_tokenize(textcorpus)
    tagged=nltk.pos_tag(tokenized)
    obj=nltk.ne_chunk(tagged)
    print(obj.draw())
for des in job_description[0:5]:
    NER(des)
```
